- The equation is as follows: **T(°F) = T(°C) × 9/5 + 32**
- For Example, let's convert 0°C celsius temperature to Fahrenheit:  **(0°C × 9/5) + 32 = 32°F**

# STEP #3: CREATE TESTING AND TRAINING DATASET

```
#X_test = Temperature_df['Celsius'].tail(6)
#y_test = Temperature_df['Fahrenheit'].tail(6)
```

```
x = Temperature_df['Celsius']
y = Temperature_df['Fahrenheit']
```

```
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, test_size=0.2)
```

# STEP#4: BUILD AND TRAIN THE MODEL

```
X_train.shape
```

```
y_train.shape
```

This will model a simple linear equation.

```
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units=1, input_shape=[1]))
```

```
model.summary()
```

```
model.compile(optimizer=tf.keras.optimizers.Adam(0.5), loss='mean_squared_error')
```

```
epochs_hist = model.fit(X_train, y_train, epochs = 100)
```

# STEP#5: EVALUATING THE MODEL

```
epochs_hist.history.keys()
```

```
plt.plot(epochs_hist.history['loss'])
plt.title('Model Loss Progress During Training')
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
plt.legend(['Training Loss'])
```

```
model.get_weights()
```

```
# Use the trained model to perform predictions

Temp_C = X_test
Temp_F = model.predict([Temp_C])
print('Temperature in degF Using Trained ANN =', Temp_F)
```

```
# Let's confirm this Using the equation:
Temp_F1 = 9/5 * Temp_C + 32
print('Temperature in degF Using Equation =', Temp_F1)
```

```
plt.plot(Temp_F, Temp_F1)
plt.title('Model Loss Progress During Testing')
plt.xlabel('Predicted')
plt.ylabel('Calculated')
plt.legend(['Variation'])
```
